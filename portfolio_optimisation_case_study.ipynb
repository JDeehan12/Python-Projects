# Case Study - Portfolio Optimisation
In this case study, I will build two $10,000 investment portfolios containing four stocks. The first portfolio will have an equal weighting between the stocks. The second portfolio will be optimised with a weighting allocation that provides the best return, adjusted for risk. To build these two portfolios, I will:
1. Import two years of data for four stocks
2. Build the initial portfolio with equal weighting to each of the stocks
3. Analyse and visualise the equal-weighted portfolio
4. Generate 10,000 portfolio scenarios with random Iighting to each of the stocks
5. Identify the optimal portfolio from the scenarios and visualise the results

## Import Packages & Connect to Data

```python
# Import packages needed for case study
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# pd.set_option('mode.chained_assignment', None)
```

```python
# List the four stock ticker symbols for my portfolio
stock_list = ['AMD', 'AAPL', 'MSFT', 'ORCL']

# Create an empty dictionary to store my stock info
stocks = {}

# Loop through each stock in the stock_list
for i_stock in stock_list:
    stocks[i_stock] = pd.read_csv(str(i_stock + '.csv'), parse_dates=True, index_col = 'Date')
```

```python
# Examine the 'AMD' Adj Close from the stocks dictionary
stocks['AMD'].head()
```

## Create the Equal-Iighted Portfolio
To create the equal-Iighted portfolio, I need to add some additional columns to the DataFrames in the `stocks` dictionary. The three columns that I will build are:
* Normalised Return = Adjusted Close / Adjusted Close on the `startdate` of the portfolio
* Allocation = Normalised Return * 0.25 (equal Iighting for each of the four stocks)
* Position Value = Allocation * 10,000 (value of the portfolio)

```python
# Create 'Normalised Return' column for each stock
for stock_name, stock_data in stocks.items():
    first_adj_close = stock_data.iloc[0]['Adj Close'] # Select the first row from the Adj Close column
    stock_data['Normalised Return'] = stock_data['Adj Close'] / first_adj_close
```

```python
stocks['AAPL'].head()
```

```python
# Create allocation for each stock - equally weighted in my initial portfolio
for stock_name, stock_data in stocks.items():
    stock_data['Allocation'] = stock_data['Normalised Return'] * 0.25
```

```python
stocks['MSFT'].head()
```

```python
# Set the value of the portfolio to $10k
for stock_name, stock_data in stocks.items():
    stock_data['Position Value'] = stock_data['Allocation'] * 10000
```

```python
stocks['ORCL'].head()
```

## Calculate Performance Metrics for the Portfolio
Now that I have created and visualised the equal-weighted portfolio, I can calculate a few metrics to further measure the performance of the portfolio. I will create five performances metrics:
 * Cumulative Return
 * Mean Daily Return
 * Standard Deviation Daily Return
 * Sharpe Ratio
 * Annualised Sharpe Ratio

```python
# Define the end and start value of the portfolio
end_value = position_values['Total'][-1]
start_value = position_values['Total'][0]

# Calculate the cumulative portfolio return as a percentage
cumulative_return = end_value / start_value - 1

print(str(round(cumulative_return*100,2)), '%')
```

```python
# Create a 'Daily Returns' column
position_values['Daily Return'] = position_values['Total'].pct_change()

position_values.head()
```

```python
# Calculate the mean Daily Return 
mean_daily_return = position_values['Daily Return'].mean()

print('The mean daily return is:', str(round(mean_daily_return, 4)))
```

```python
# Calculate the standard deviation of Daily Return 
std_daily_return = position_values['Daily Return'].std()

print('The std daily return is:', str(round(std_daily_return, 4)))
```

### Sharpe Ratio
Now, I'll explore a risk-adjusted return metric called the sharpe ratio. The sharpe ratio helps us quantify how much return I am getting by a given level of risk. When comparing two different investments against the same benchmark, the asset with the higher Sharpe ratio provides a higher return for the same amount of risk or the same return for a lower risk than the other asset. 

It is calculated by the average return of the portfolio minus a risk free rate (such as government bonds), divided by the standard deviation of the return. In this case, I assume the risk-free rate is close to zero, so I won't add it to the formula.

```python
# Calculate the sharpe ratio
sharpe_ratio = mean_daily_return / std_daily_return

sharpe_ratio
```

```python
# Calculate the annualised sharpe ratio
sharpe_ratio_annualised = sharpe_ratio * 252**0.5

sharpe_ratio_annualised
```

## Prepare Scenarios to Optimise Portfolio Weighting
I need to prepare my data ahead of generating my scenarios to optimise the portfolio weighting. I will:
 * Create a dictionary containing the adjusted close for each of my stocks: stock_adj_close
 * Create another dictionary that transforms the adjusted close for each day to a percent change from the previous day

```python
# Create stock_adj_close dictionary
stock_adj_close = {}

for stock_name, stock_data in stocks.items():
    stock_adj_close[stock_name] = stock_data['Adj Close']
```

```python
# Convert the stock_adj_close dictionary to a DataFrame
stock_adj_close = pd.DataFrame(data=stock_adj_close)

stock_adj_close.head()
```

```python
# Create stock_returns DataFrames to see the day over day change in stock value
stock_returns = stock_adj_close.pct_change()

stock_returns.head()
```

## Build & Run 10,000 Portfolio Scenarios
Now that I've prepared my data, I'm almost ready to run my scenarios. First, I need to build the structures required to generate these scenarios and store the output. To do this, I will use the `numpy.zeros()` function. 

This function creates arrays that are filled with zeros. After I run the scenarios, I replace these zeros with the corresponding output. The reason I create the arrays with zeros first is to give my arrays the correct shape before I replace them with the correct values.

I will create four different arrays:
 * weights_array - this array will have 10,000 rows and 4 columns and hold the weighting allocation for each stock
 * returns_array - this array will contain the portfolio return for each scenario
 * volatility_array - this array will contain the portfolio volatility for each scenario
 * sharpe_array - this array will contain the sharpe ratio for each scenario

```python
# Define the number of scenarios and create a blank array to populate stock weightings for each scenario
scenarios = 10000

weights_array = np.zeros((scenarios, len(stock_returns.columns)))

weights_array
```

```python
# Create additional blank arrays for scenario output
returns_array = np.zeros(scenarios)
volatility_array = np.zeros(scenarios)
sharpe_array = np.zeros(scenarios)
```

```python
import random
random.seed(3)
np.random.seed(3)

for index in range(scenarios): 
    # Generate four random numbers for each index
    numbers = np.array(np.random.random(4))
    
    # Divide each number by the sum of the numbers to generate the random weight
    weights = numbers / np.sum(numbers)
    
    # Save the weights in weights_array
    weights_array[index,:] = weights
    
    # Calculate the return for each scenario
    returns_array[index] = np.sum(stock_returns.mean()*252*weights)
    
    # Calculate the expected volatility for each scenario
    volatility_array[index] = np.sqrt(np.dot(weights.T,np.dot(stock_returns.cov()*252, weights)))

    # Calculate the Sharpe Ratio for each scenario 
    sharpe_array[index] = returns_array[index] / volatility_array[index]
```

```python
print("The first combination:", weights_array[0])
```

```python
print("The sharpe ratio of the first portfolio:", sharpe_array[0])
```

## Identify the Optimal Portfolio
Now that I have the output for all 10,000 scenarios, I can identify the optimal portfolio. The optimal portfolio in this case study is the portfolio that has the highest sharpe ratio.

```python
# Find the highest sharpe ratio in sharpe_array
sharpe_array.max()
```

```python
# Find the index of the optimal portfolio
index_max_sharpe = sharpe_array.argmax()
index_max_sharpe
```

```python
# Print the optimal weights for each stock
print(stock_list)
print(weights_array[index_max_sharpe,:])
```

## Visualise the Optimal Portfolio & Portfolio Scenarios
Now I'll visualise my portfolio scenarios by using a scatter chart. I can use the volatility and returns arrays on each axis to see the relationship between risk and reward. As a final step, Ican visualise where the optimal portfolio appears among all of the scenarios.

```python
# Visualise volatility vs returns for each scenario
plt.figure(figsize=(12,8))

plt.scatter(volatility_array, returns_array, c=sharpe_array, cmap='viridis')

plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return');
```

```python
# Identify the optimal portfolio in the returns and volatility arrays
max_sharpe_return = returns_array[index_max_sharpe]
max_sharpe_volatility = volatility_array[index_max_sharpe]

# Visualise volatility vs returns for each scenario
plt.figure(figsize=(12,8))

plt.scatter(volatility_array, returns_array, c=sharpe_array, cmap='viridis')

plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')

# Add the optimal portfolio to the visual
plt.scatter(max_sharpe_volatility, max_sharpe_return, c='orange', edgecolors='black');
```
